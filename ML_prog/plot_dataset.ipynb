```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("G:\\ML\\student.csv")
print(df)

plt.scatter(df.index, df["Exam Score"], color='blue')
plt.xlabel("Student Index")
plt.ylabel("Exam Score")
plt.title("Scatter Plot of Student Exam Scores")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("G:\\ML\\student.csv")
print(df)

plt.bar(df["Student Name"], df["Exam Score"], color='green')
plt.xlabel("Student Name")
plt.ylabel("Exam Score")
plt.title("Bar Plot of Student Exam Scores")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("G:\\ML\\student.csv")
print(df)

plt.pie(df["Exam Score"],
        labels=df["Student Name"],
        autopct="%1.1f%%",
        startangle=90)

plt.title("Pie Chart: Exam Score Distribution")
plt.show()
```
